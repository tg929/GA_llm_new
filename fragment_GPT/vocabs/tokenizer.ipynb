```python
from deepchem.feat.smiles_tokenizer import SmilesTokenizer

# 创建 SMILES Tokenizer 实例  vocab.txt download from https://github.com/deepchem/deepchem/blob/2.4.0/deepchem/feat/tests/data/vocab.txt
tokenizer = SmilesTokenizer("./vocab.txt")  

# 获取词表（vocab）
vocab = tokenizer.vocab

# 打印词表
print("Tokenizer Vocabulary:")
for token, idx in vocab.items():
    print(f"Token: {token}, Index: {idx}")
```

```python
# 测试单个字符 '|'
encoded_pipe = tokenizer.encode('|')
decoded_pipe = tokenizer.decode(encoded_pipe)
print(f"Encoded '|': {encoded_pipe}")
print(f"Decoded from {encoded_pipe}: {decoded_pipe}")

# 测试特殊标记 [SEP]
encoded_sep = tokenizer.encode('[SEP]')
decoded_sep = tokenizer.decode(encoded_sep)
print(f"Encoded '[SEP]': {encoded_sep}")
print(f"Decoded from {encoded_sep}: {decoded_sep}")
```

```python
# 测试单个字符 '|'
encoded_pipe = tokenizer.encode('|', add_special_tokens=False)
decoded_pipe = tokenizer.decode(encoded_pipe)
print(f"Encoded '|': {encoded_pipe}")
print(f"Decoded from {encoded_pipe}: {decoded_pipe}")

# 测试特殊标记 [SEP]
encoded_sep = tokenizer.encode('[SEP]', add_special_tokens=False)
decoded_sep = tokenizer.decode(encoded_sep)
print(f"Encoded '[SEP]': {encoded_sep}")
print(f"Decoded from {encoded_sep}: {decoded_sep}")
```

```python
# 测试单个字符 '|'
encoded_pipe = tokenizer.encode('[EOS]')
decoded_pipe = tokenizer.decode(encoded_pipe)
print(f"Encoded '|': {encoded_pipe}")
print(f"Decoded from {encoded_pipe}: {decoded_pipe}")

# 测试特殊标记 [SEP]
encoded_sep = tokenizer.encode('[BOS]')
decoded_sep = tokenizer.decode(encoded_sep)
print(f"Encoded '[SEP]': {encoded_sep}")
print(f"Decoded from {encoded_sep}: {decoded_sep}")
```

```python
src = 'C12OC(=O)NC(O)(C(OC)C=CC=C(C)Cc3cc(c(Cl)c(OC)c3)N(C)C(=O)CC(*)C3(C)C(O3)C1C)C2.C(CCCC(=O)Nc1nc(CN(Cc2cc(CN(Cc3nc(CC(CCCCC)=O)ccc3)Cc3ccccn3)cc(OC*)c2)Cc2ccccn2)ccc1)C'
tgt = 'CCCCCC(=O)Cc1cccc(CN(Cc2cc(CN(Cc3ccccn3)Cc3cccc(NC(=O)CCCCC)n3)cc(OCCCCC(Cc3ccc(C(=O)NCCOCCOCCN(Cc4ccc(C(=O)SCCC(=O)N(C)[C@H](C)C(=O)O[C@H]5CC(=O)N(C)c6cc(cc(OC)c6Cl)C/C(C)=C/C=C/[C@@H](OC)[C@@]6(O)C[C@H](OC(=O)N6)[C@@H](C)[C@H]6O[C@]56C)cc4)C(=O)C4(c5ccc(Cl)cc5)CCCCC4)cc3)C(=O)C3(c4ccc(Cl)cc4)CCCCC3)c2)Cc2ccccn2)n1'
```
