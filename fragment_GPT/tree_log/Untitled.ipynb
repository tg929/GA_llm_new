```python
import json

#  'data.json' JSON
file_path = 'buffer.json'

# JSON
with open(file_path, 'r', encoding='utf-8') as file:
    # json.load()Python
    data = json.load(file)

# data
data 
```

```python
# 
total_length = 0
path_count = 0

# 
for key in data.keys():
    #  = '-'  'root'  '-'
    length = key.count('-') - 1  #  'root'  '-'
    total_length += length
    path_count += 1

# 
average_length = total_length / path_count

# 
print(f" root : {average_length}")
```
