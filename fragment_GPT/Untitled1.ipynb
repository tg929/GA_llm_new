```python
import pandas as pd
import glob
```

```python
csv_files = glob.glob('./output/mcts_v7/mcts_5ht1b_v7_*.csv') 
print(len(csv_files))
csv_files
```

```python
dfs = []
for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)
# 竖向拼接所有数据框
combined_df = pd.concat(dfs, ignore_index=True)
combined_df
```

```python
# 判断 cur_sentence 中是否包含 [EOS]
combined_df['contains_EOS'] = combined_df['cur_sentence'].str.contains(r'\[EOS\]', regex=True)
```

```python
combined_df['contains_EOS']
```

```python
# 2. 按 rv 字段从大到小排序
combined_df.sort_values(by='rv', ascending=False, inplace=True)
combined_df
```

```python
# 筛选条件：rq > 0.5，rs > (10 - 5) / 9（即 rs > 5/9），rv > 10.0
filtered = combined_df[(combined_df['rq'] > 0.5) &
                       (combined_df['rs'] > (10 - 5) / 9) &
                       (combined_df['rv'] > 8.7845)]
len(filtered)
```

```python
filtered
```

```python
# 3. 计算 top 5% 数据的行数
row_count = len(combined_df)
top_5pct_count = int(0.05 * row_count)
top_5pct_count
```

```python
# 4. 取出 rv 值最高的 5% 数据
top_5pct_df = filtered[:top_5pct_count]

# 5. 计算均值和方差（这里只针对 rv 字段进行演示）
rv_mean = top_5pct_df['rv'].mean()
rv_variance = top_5pct_df['rv'].var()  # 默认是样本方差 (ddof=1)
print(f"rv最高5%的均值: {rv_mean}")
print(f"rv最高5%的方差: {rv_variance}")
```

```python
top_5pct_df.to_csv("./output/mcts_v7/mcts_5ht1b_v7_top5%.csv")
top_5pct_df
```
